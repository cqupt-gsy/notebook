### SQL-DML：增、删、改、查表中的数据
* SQL执行顺序
    * FORM: 对FROM的左边的表和右边的表计算笛卡尔积。产生虚表VT1
    * ON: 对虚表VT1进行ON筛选，只有那些符合<join-condition>的行才会被记录在虚表VT2中
    * JOIN： 如果指定了OUTER JOIN（比如left join、 right  join），那么保留表中未匹配的行就会作为外部行添加到虚拟表VT2中，产生虚拟表VT3
    * WHERE：对虚拟表VT3进行WHERE条件过滤。只有符合<where-condition>的记录才会被插入到虚拟表VT4中
    * GROUP BY: 根据group by子句中的列，对VT4中的记录进行分组操作，产生VT5
    * HAVING：对虚拟表VT5应用having过滤，只有符合<having-condition>的记录才会被插入到虚拟表VT6中
    * SELECT：执行select操作，选择指定的列，插入到虚拟表VT7中
    * DISTINCT：对VT7中的记录进行去重。产生虚拟表VT8
    * ORDER BY: 将虚拟表VT8中的记录按照<order_by_list>进行排序操作，产生虚拟表VT9
    * LIMIT：取出指定行的记录，产生虚拟表VT10, 并将结果返回
* 数据库的查询：
    * SELECT FROM WHERE
        * FROM中指定的TABLE会嵌套循环（以行为单位），组成新的表提供给WHERE作为输入源
        * WHERE过滤FROM中的结果集
            * WHERE中的BETWEEN是闭区间
            * 可以使用元组比较：A.ID = B.ID AND A.C = B.C 等价于(A.ID, A.C) = (B.ID, B.C)
        * SELECT选择需要的列
    * ORDER BY：默认使用升序，DESC表示降序，ASC表示升序
    * GROUP BY：按照列元素分组，在WHERE之后执行，先分组然后再执行聚集函数
        * 出现在SELECT中的列，如果没有出现在聚集函数中，必须全部出现在GROUP BY中
        * 如果没指定GROUP BY，则整个结果集被分成一个组
    * HAVING：对分组结果的过滤，可以使用聚集函数，在GROUP BY之后执行
        * 出现在HAVING中的列，如果没有出现在聚集函数中，必须全部出现在GROUP BY中
    * 子查询：
        * WHERE中的子查询：
            * IN|NOT IN：列值在或者不在某个子集合结果中，返回结果子集
            * < SOME|<= SOME|> SOME|>= SOME|= SOME|<> SOME：列值至少比集合中某些值小|小于等于|大于|大于等于|等于|不等于，其中= SOME等价于IN
            * < ALL|<= ALL|> ALL|>= ALL|= ALL|<> ALL：列值比集合中的所有值小|小于等于|大于|大于等于|等于|不等于，其中<> ALL等价于NOT IN
            * EXISTS|NOT EXISTS：测试子集合中的元素是否存在，返回TRUE或FALSE
            * UNIQUE|NOT UNIQUE：测试子集合中的元素是否重复，返回TRUE或FALSE
        * FROM中的子查询：
            * FROM中的子查询结果做为新表提供给外层查询，可以给该新表以及表中的列创建别名方便外层查询使用
            * LATERAL关键字可以让后面的子查询看到前面子查询的变量范围
        * WITH AS：定义临时表，当子查询晦涩难懂时，可以先用WITH AS创建相关临时表
        * SELECT中的子查询：子查询只返回一列的结果集只能出现在SELECT, WHERE, HAVING中
    * 连接查询
        * 组合查询如果不限制WHERE条件谓词，结果集就是两个表的笛卡尔乘积
        * 自然连接（内连接）：只选择两个表中属性相同并且值相等的集合，也就是组合查询中指定WHERE条件谓词，例如如下四种等价的形式：
            * WHERE A.id = B.id
            * FROM A NATURAL JOIN B：不指定两个表必须相等的列，则执行引擎会把所有列名相同的数据进行对比
            * FROM A [NATURAL] JOIN B USING (column_name)：USING指定两个连接表必须相等的列
            * FROM A [INNER] JOIN B ON A.id=B.id：ON相当于SELECT中的WHERE关键字，可以写很复杂的条件谓词
            * 注意：在FROM中指定自然连接会直接产生新的表并作为WHERE的输入源；而FROM中不指定自然连接，则会将两个表的笛卡儿积作为WHERE的输入源
        * 外连接，避免自然连接中数据丢失（自然连接中如果出现不匹配的数据则不会显示，但是外连接会保留），先执行内连接，然后根据不同的外连接把丢失的数据补上
            * 左外连接（FROM A [NATURAL] LEFT OUTER JOIN B）：保留左外连接左边关系中的元组，保留A中的元组
            * 右外连接（FROM A [NATURAL] RIGHT OUTER JOIN B）：保留右外连接右边关系中的元组，保留B中的元组
            * 全外连接（FROM A [NATURAL] FULL OUTER JOIN B）：保留出现在两个关系中的元组，保留A和B中的元组
    * 集合运算：每个SELECT从句是一个集合，集合运算是应用在多个集合上的运算，且默认去重
        * UNION|UNION ALL：求集合之间的并集
        * INTERSECT|INTERSECT ALL：求集合之间的交集；去重操作会保留集合中出现次数较少的
        * EXCEPT|EXCEPT ALL：求集合之间的差集；保留集合A中出现且不在集合B中出现的数据集；去重操作按集合A中数据出现的次数减去B中数据出现的次数结果处理
    * AS应用在SELECT中将修改输出结果的列名；AS应用在FROM中将创建表的别名
    * %：匹配任意子串，_：匹配任意一个字符，模式匹配的字符串是大小写敏感的
    * NULL：
        * WHERE谓词中处理时，NULL比较结果是UNKNOWN；TRUE OR UNKNOWN结果返回TRUE，其他都是返回UNKNOWN
        * DISTINCT处理NULL时，NULL=NULL结果返回TRUE
    * 普通聚集函数：AVG, MIN, MAX, SUM, COUNT
    * 高级聚集函数：RANK() OVER(COLUMN),  NTILE(N)
* 数据库的修改：
    * DELETE FROM WHERE
        * FROM指定需要被删除的表数据
        * WHERE条件过滤FROM中的数据，满足条件的数据将会被删除
        * 每个DELETE关键字只能指定一个表
    * INSERT INTO
        * 如果不指定需要插入的列数据，则VALUES里面的列数据需要和表定义时列的顺序一致
        * INSERT INTO SELECT，先执行完SELECT，再执行INSERT
    * UPDATE SET
        * WHERE条件过滤需要被修改的数据
        * CASE WHEN THEN ELSE多条件判断设定值
        
### SQL-Integrity：定义表中列的完整性约束
* 单个表上的约束，主要是约束列：NOT NULL, UNIQUE, CHECK(<WHERE中的条件判断>)
* 主键约束，PRIMARY KEY([COLUNM_NAME])
* 外键约束，FOREIGN KEY([COLUNM_NAME]) REFERENCES TABLE_NAME
    * ON DELETE CASCADE：指定删除级联操作
    * ON UPDATE CASCADE：指定更新级联操作
    * 级联操作表示：如果对主表的相关数据修改将会触发从表对应列数据的更改
* 完整性约束默认是立即执行，也就是说更新操作一旦开始就会执行完整性约束检查
    * INITIALLY DEFERRED：加到约束的声明中表示该约束可以延迟检查
    * SET CONSTRAINTS [constraints-list] DEFERRED：在事务中指定相关约束延迟检查，保证不在事务中间检查相关的约束
* 复杂的CHECK条件开销很大，所有涉及的表发生变化，CHECK都要执行，建议使用触发器来完成对应的CHECK操作

### SQL-Authorization：数据库授权相关操作
* GRANT ON TO：
    * SELECT：查询权限
    * UPDATE(列名)：更新权限，如果指定列名，则只能更新对应的列
    * INSERT(列名)：插入权限，如果指定列名，则只能插入对应的列，其他列要么为NULL要么为默认值
    * DELETE：删除权限
    * PUBLIC：如果用户名指定为PUBLIC，则系统当前用户和将来的用户都会被授权
    * 创建表的用户，默认拥有该表的所有操作权限
    * WITH GRANT OPTION：允许被授权的用户继续将权限授予给别的用户
* REVOKE ON FROM：与授权操作是反向操作
    * CASCADE：默认权限收回规则，级联收回
    * RESTRICT：禁止级联收回权限
* CREATE ROLE：创建角色，类似Linux系统中的用户组；可以授权给角色，在把角色授权给用户；角色的权限是可以继承的，例如将角色A授予给用户时，该用户就拥有了该角色的权限

# SQL执行计划
* 查询语句->语法分析器->关系代数表达式->计算原语（关系代数加上执行注释）->查询执行计划（执行一个特定的计算原语）->查询执行引擎（执行执行计划）
* 选择最优的执行计划执行的过程叫做查询优化；查询优化器执行逻辑：产生等价的表达式，对表达式做注释产生查询计划，估计执行计划的代价
* 等值比较会选择HASH索引查找的原因是查找速度优于顺序查找，范围查找不使用HASH索引的原因是因为找到边界值后，顺序读取文件剩下的部分就是结果集
* EXPLIAN QUERY：查看执行计划

# 索引与散列
* 索引的创建和删除：
    * CREATE INDEX INDEX_NAME ON TABLE_NAME (COLUNMS)
    * DROP INDEX INDEX_NAME
* 顺序索引
    * 聚集索引：数据的物理存储顺序与索引列指定的顺序一致则为聚集索引；大多数情况都是按主键顺序存储，所以索引列是主键的一般都是聚集索引（大部分数据库会自动为主键建立一个聚集索引，也就是表中的数据按主键的顺序存放）
    * 非聚集索引：数据的物理存储顺序与索引列指定的顺序不一致则为非聚集索引，必须是稠密索引
    * 索引项：由一个索引列值和指向对应记录的指针组成
        * 稠密索引：每个索引项都对应一条记录；其结构类似一个数组指针，对应的数据记录物理存储是无序的，因此具有相同索引列值的记录顺序链式存储在第一条记录之后
        * 稀疏索引：只有部分索引项有对应的记录，只能是聚集索引；其结构类似一个链表，对应的数据记录物理存储是顺序存储的
    * 索引项太多，不能全部加载进入内存的弊端：会产生多次的磁盘读，每次将数据从磁盘读入内存是非常耗时的。使用B+树建立索引可以完美解决这个问题，思想就是建立多级索引，每一级的索引项（B+树的一个节点）都足够放入内存，访问磁盘的次数就折算为了B+树的高度
    * 顺序索引匹配规则：
        * 使用多个单列索引：使用一个列索引找出相关记录，然后用条件谓词过滤其他的记录；同时使用所有列索引找出对应记录，然后计算集合的交集
        * 使用多列索引：多列索引符合最左匹配原则，且只支持等值比较，一旦出现范围查询，在范围查询以后的条件谓词都不会应用索引搜索了
* 散列索引
    * 俗称HASH索引，根据索引列的值计算其HASH值，将该HASH值以及记录指针构造成索引项
    * 瓶颈在于HASH索引碰撞，也就是多个列值计算出来的HASH值相同发送冲突，可以使用链式结构顺序连接对应的数据记录指针
    * 范围查询顺序索引更优（聚集索引），等值查询散列索引更优
* 位图索引
    * 一个大数组，计算列值的数值，然后以该数值作为数组的下标，并把该下标对应的元素标记为1，表示为该记录，100万位的位图大小才128KB左右，比HASH更节省空间

# SQL-Transaction control：操作数据库事务
* 事务本质理解就是多线程读写数据时的同步问题
* 事务四大特性ACID：
    * 原子性：一个事务是一个最小的不可分割的操作单元，要么全部成功要么全部失败
    * 隔离性：一个事务在最终提交前，对其他事务是不可见的
    * 一致性：数据库总是从一个一致性状态转变为另一个一致性状态
    * 持久性：一旦事务被提交，其修改永久性的保存到数据库中
* 事务的隔离级别：
    * READ UNCOMMITTED：允许读取未提交的数据，存在脏读，不可重复读，幻读等问题
    * READ COMMITTED：只允许读取已提交的数据，解决脏读的问题，存在不可重复读，幻读等问题
    * REPEATABLE READ：只允许读取已提交的数据，并且在一个事务两次读取一个数据的期间不允许其他事务对该数据进行更新，解决不可重复读的问题，存在幻读等问题
    * SERIALIZABLE：串行化，保证事务是串行化执行，解决幻读的问题
    * 所有事务的都不允许脏写，也就是数据被未提交的事务占有时，其他事务不允许写
    * 不可重复读：在事务中两次读取同一个数据的结果不一致，期间有其他事务改变了该数据并且完成了提交
    * 幻读：在事务两次读取数据时，数据数量发生变化，期间有其他事务插入了新数据
* 事务隔离级别的实现，并发控制机制：
    * 悲观的控制机制：基于锁的协议和基于时间戳的协议，操作开始之前就会锁定资源
    * 乐观的控制机制：多版本时间戳机制和快照隔离，并不会明确的锁定资源，或者只在最终提交时请求锁资源
    * 基于锁的协议：
        * 共享锁：获得某数据项共享锁的事务，只能读该数据项但是不能写
        * 排他锁：获得某数据项排他锁的事务，既可以读也可以写该数据项
        * 意向锁：在给数据加共享锁或者排他锁时，需要从锁粒度树根节点开始遍历直到找到该数据的粒度节点，在遍历过程中经历的所有节点都会加上意向锁，主要是为了更快的知道当前数据集中包含哪些锁以及当前事务需要哪些锁
        * 锁管理器：维护一个链表（成为锁表），链表的节点为已经加锁的数据，每个数据包含一个已授予锁或者等待锁的事务列表
            * 当发起一个锁请求时，数据在锁表中已经存在，则会在该数据的事务列表末端添加一个新节点
            * 当数据在链表中不存在时，如果该事务没有在其他数据中获得锁或者等待锁的情况下，会直接把锁授予给该事务，否则会让该事务等待锁资源，并将该数据记录到锁表中
            * 当收到事务释放锁请求时，会在对应数据的事务列表中删除该事务节点，其次会遍历其他所有的数据上该事务的请求锁状态，并依次处理
            * 如果事务被终止，则会遍历数据中所有该事务请求锁资源的节点
        * 死锁处理：
            * 死锁预防：保证加锁请求的顺序；根据锁请求事务的时间优先选择事务回滚
            * 死锁检测：利用有向图维护事务的关系，当图出现环就代表有死锁，需要回滚事务
    * 基于时间戳的协议：根据时间戳决定事物的执行顺序
        * 事物开始执行时分配事物对应的时间戳，可以是系统时间也可以是自增型变量
        * 在事务中每次读完或者写完对应的数据，将事务的时间戳分配给该数据的读写时间戳
        * 事务发出读请求时：
            * 如果该事务的时间戳在数据的写时间之前，则表明已经有其他事务重写过该数据了，因此不可读
            * 如果该事务的时间戳在数据的写时间之后，则表明该写操作是最新的，可以读，并且更新该数据的读时间戳
        * 事务发出写请求时：（Thomas写规则就是在写操作不可写时，忽略对应事务的写操作而不是回滚事务）
            * 如果该事务的时间戳在数据的读时间之前，则说明该写操作已经过时，因此不可写
            * 如果该事务的时间戳在数据的写时间之前，则说明其他事务已经写过，并且该事务的写操作以及过时，因此不可写
            * 如果该事物的时间戳在数据的写时间之后，则允许写操作，并且更新该数据的写时间为该事务的时间戳
    * 多版本时间戳机制
        * 每个数据项包含内容值，写时间戳，读时间戳三个字段，代表着一个版本
        * 当事务新写一个数据时，会记录对应的值，并将读、写时间戳记录为该事务的时间戳
        * 读操作与基于时间戳的协议类似，但是不会发生阻塞，如果读事务的时间在数据的写时间戳之前，会选择读取历史版本
        * 更新操作与基于时间戳的协议类似，但是更新完后会有一个更新版本号额外的操作
    * 快照版本隔离
        * 事务在自己的私有空间操作，只在最终提交时才判断冲突
        * 先提交者获胜策略
            * 提交时检查是否有其他事务以及对数据进行了提交，如果有则事务终止，如果没有事务则可以提交
        * 先更新者获胜策略
            * 提交时去获取对应数据项的写锁，如果被其他事务占有则等待；当获得锁后判断是否有事务已经对该数据进行了更新，如果是则事务终止，否则事务提交
* 补偿事务：执行事务的反向操作，包括执行顺序也要相反
    * SET [GLOBAL|SESSION] TRANSACTION ISOLATION LEVEL READ CIMMITED|REPEATABLE READ|SERIALIZABLE：改变事务的隔离级别，并且必须出现在事务最开始

# 数据库的恢复
* 系统日志内容包括：事务ID，数据行标识，旧的列值，新的列值，事务开始事件，事务提交事件，事务终止事件等，事务执行写操作之前会将该日志记录持久化到日志文件中
    * redo：根据日志将数据行的值都替换为新的列值
        * 如果存在事务开始事件，同时存在事务提交或者事务终止事件
    * undo：根据日志将数据行的值都替换为旧的列值
        * 如果存在事务开始事件，不存在事务提交和事务终止事件

# 分布式数据库
* 数据库划分：
    * 水平划分：指同一表中不同的行记录分配到不同的物理文件中，包括范围划分，HASH划分，轮训划分
    * 垂直划分：指同一表中不同的列分配到不同的物理文件中
* 分布式事务：
    * 事务管理器：管理单个节点上数据库的ACID
    * 事务协调器（集中式架构只有中心节点需要）：协调多个节点之间的事务生命周期
    * 分布式事务故障：站点故障，消息丢失，通信链路故障，网络划分
    * 分布式事务一致性保障：
        * 两阶段提交协议：（存在表锁死，导致本地数据库事务也无法进行，并且不好与第三方开展对应的事务处理）
            * 所有应该执行事务的站点执行完事务后，通知发起站点的事务协调器，开始2PC
            * 事务协调器记录```<Prepare T>```到日志，并将该事件发送到所有站点，其他站点收到消息后，如果能提交则记录```<Ready T>```到日志，并将该事件回复给发起站点的事务协调器；否则记录```<No T>```到日志，并发送给发起站点的事务协调器
            * 发起站点的事务协调器接收到其他站点的消息后，如果所有回复都是```<Ready T>```，则记录```<Commit T>```到日志，并将事件发送给其他站点做写日志操作和最终的提交操作；否则记录```<Abort T>```到日志，并将事件发送给其他站点做写日志操作和终止操作
        * 两阶段提交协议故障恢复：
            * 参与站点故障：
                * 参与站点在发送```<Ready T>```之前发生故障，默认该站点返回```<No T>```
                * 参与站点在发生```<Ready T>```之后发生故障
                    * 如果日志包含```<Commit T>```，则执行该站点的redo操作
                    * 如果日志包含```<Abort T>```，则执行该站点的undo操作
                    * 如果日志只有```<Ready T>```，则该站点需要咨询其他的站点关于该事件的最终结果并执行对应的redo或者undo操作
                    * 如果没有任何上述事件，则执行undo操作
            * 协调器故障：
                * 其他站点日志中包含```<Commit T>```，则协调器站点必须提交
                * 其他站点日志中包含```<Abort T>```，则协调器站点必须终止
                * 其他某些站点日志中不包含```<Ready T>```，则协调器站点必须终止
                * 其他站点日志中只有```<Ready T>```，必须等待协调器恢复，引起阻塞（可以在记录事务事件的同时记录起拥有的锁，避免将额外的数据锁死导致数据库服务不可用）
        * 持久化消息，消息队列
            * 发送站点：
                * 将消息写入专门的关系表中
                * 消息发送进程监控发送消息表，有消息就发送；当且仅当发送站点的事务提交后，进程才可以读取发送消息表中的消息，如果事务终止，则会删除发送消息表中的消息
                * 消息发送进程收到接收站点的回复后才会删除对应的消息，如果没收到回复则会重试一段时间，否则就会让异常处理代码执行补偿事务
            * 接收站点：
                * 接收到消息后，会将消息写入消息接收表中
                * 事务提交后或者该消息已经存在，会发送确认回复消息给发送站点
* 分布式数据库并发控制
    * 分布式锁管理：集中式，统一去一台中心节点申请锁；分布式，通过多个节点共享锁信息
        * 全局等待树，查找树中是否存在环判断死锁
    * 分布式时间戳管理：集中式，统一去一台中心节点申请时间戳；分布式，通过本地时间戳加节点名称构造全局时间戳
* CAP定理，分布式只能满足以下三个性质中的两个：
    * 一致性：分布式中所有节点的数据状态是一致的，外界查询分布式集群时，结果永远一样
    * 可用性：分布式中，如果部分节点失效，该集群还是可以正常工作
    * 分区容忍性：容忍分布式集群发生网络分区
* BASE性质，对强一致性要求不是太高的系统的一种妥协选择
    * 基本上可用：最高需求，牺牲一致性
    * 软状态：指数据库在某段时间可能存在不确定性
    * 最终一致性：保证数据的状态最终是一致的
* 分布式存储系统：Bigtable, GFS, HBase
* 长事务，嵌套事务，多级事务

# SQL触发器
* CREATE TRIGGER TRIGGER_NAME AFTER/BEFORE INSERT/UPDATE/DELETE ON TABLE
* REFERENCING NEW ROW AS NEW_ROW_NAME：将新修改的行数据赋值给NEW_ROW_NAME
* REFERENCING OLD ROW AS OLD_ROW_NAME：将被改变的行数据赋值给OLD_ROW_NAME
* REFERENCING NEW TABLE AS NEW_TABLE_NAME：将被修改后的表内容赋值给NEW_TABLE_NAME，且只能用于AFTER事件的触发器
* REFERENCING OLD TABLE AS OLD_TABLE_NAME：将被修改前的表内容赋值给OLD_TABLE_NAME
* FOR EACH ROW：迭代每行被修改的数据，并分别赋值到REFERENCING NEW ROW和REFERENCING OLD ROW中
* FOR EACH STATEMENT：针对每条语句做相应处理
* WHEN CONDITION AND CONDITION BEGIN END：只有WHEN中的条件满足，才会执行BEGIN END之间的语句
* ALTER TRIGGER TRIGGER_NAME DISABLE：关闭触发器
* DROP TRIGGER TRIGGER_NAME：删除触发器

# 数据库范式
* 第一范式(1FN)：关系中的属性都不可再分，属性是原子性的，RDBMS中存在的表都符合1FN
    * 通俗的说：字段是最小的单元不可再分，即列不能够再分成其他几列；不能是复合属性、不能是多值属性、不能是重复属性；
    * 存在大量的数据冗余
    * 增、删、改存在异常
* 第二范式(2FN)：在1FN的基础上，消除非主属性对于码的部分函数依赖
    * 通俗的说：表中的字段必须完全依赖于全部主键而非部分主键，即表必须有一个主键，没有包含在主键中的列必须完全依赖于主键，而不能只依赖于主键的一部分；表中必须存在侯选关键字，即每一行不同于其他任一行，是惟一区分的，任何非关键字段不能依赖于侯选关键字的一部分
    * 消除数据冗余，解决修改异常
    * 增、删存在异常
* 第三范式(3FN)：在2FN的基础上，消除非主属性对于码的传递函数依赖
    * 通俗的说：非主键外的所有字段必须互不依赖，非关键字字段必须直接依赖任一侯选关键字，也就是说表中不能含有其他表的非主属性
    * 消除大量数据冗余，解决修改异常
    * 解决增、删异常
* BC范式：在3NF的基础上，消除主属性对于码的部分与传递函数依赖，没有任何属性完全函数依赖于非候选码的任何一组属性
    * 通俗的说：每个表中只有一个候选键
* 第四范式(4FN)：在BCNF的基础上，消除表中的多值依赖
